```python
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['Times New Roman', 'SimSun']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号

# 时间节点与对应事件（简写以适配图像空间）
years = [
    1998, 2000, 2004, 2008, 2012, 2014,
    2016, 2017, 2018, 2021, 2022, 2025
]
events = [
    "PAGSI计划启动",
    "政府门户Service-Public.fr上线",
    "ADELE电子政务计划",
    "France Numérique 2012",
    "设立SGMAP改革协调机构",
    "国家平台战略 + CDO任命",
    "France-Connect统一认证平台",
    "PPNG计划 + Action Publique 2022改革",
    "AI国家战略（AI for Humanity）",
    "France 2030，AI战略第二阶段",
    "国务法院主张AI治政战略",
    "AI行动峰会 + 国际宣言签署"
]

# 绘图
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_ylim(-1, 1)
ax.set_xlim(min(years) - 1, max(years) + 1)

# 画横轴
ax.hlines(0, min(years), max(years), color='gray', linewidth=1)

# 标注节点
for i, (year, event) in enumerate(zip(years, events)):
    ax.plot(year, 0, 'o', color='black')
    va = 'bottom' if i % 2 == 0 else 'top'
    offset = 0.1 if i % 2 == 0 else -0.1
    ax.text(year, offset, f"{year}\n{event}", ha='center', va=va, fontsize=9)

# 清除坐标轴
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 法语版事件列表
events_fr = [
    ("1998", "Programme PAGSI"),
    ("2004", "ADELE (e-administration)"),
    ("2008", "France Numérique 2012"),
    ("2012", "SGMAP"),
    ("2014", "Stratégie État plateforme"),
    ("2016", "France-Connect"),
    ("2017", "Plan Action Publique 2022"),
    ("2018", "Stratégie nationale IA"),
    ("2021", "France 2030"),
    ("2022", "IA pour la gouvernance"),
    ("2025", "AI Action Summit")
]

# 总统任期（起始位置，跨度），标签
presidents = [
    (0, 2, "Jacques Chirac\n(1995–2007)"),
    (2, 1, "N. Sarkozy\n(2007–2012)"),
    (3, 2, "F. Hollande\n(2012–2017)"),
    (5, 6, "E. Macron\n(2017–présent)")
]

# 创建画布
fig, ax = plt.subplots(figsize=(12, 2.5))
ax.set_ylim(-1.2, 0.6)
ax.set_xlim(-0.5, len(events_fr)-0.5)

# 时间轴横线
ax.hlines(0, 0, len(events_fr)-1, color='black', linewidth=1)

# 绘制事件节点
for i, (year, label) in enumerate(events_fr):
    ax.plot(i, 0, 'o', color='black')
    va = 'bottom' if i % 2 == 0 else 'top'
    offset = 0.15 if i % 2 == 0 else -0.15
    ax.text(i, offset, f"{year}\n{label}", ha='center', va=va, fontsize=10.5)

# 添加总统任期横条
for start, width, name in presidents:
    ax.broken_barh([(start, width)], (-0.9, 0.3), facecolors='lightgray')
    ax.text(start + width / 2, -0.75, name, ha='center', va='center', fontsize=9)

# 去除坐标轴
ax.axis('off')

# 保存 SVG
svg_path = "./france_ai_timeline_presidents.svg"
plt.savefig(svg_path, format='svg')
plt.show()

svg_path
```
